```python
import math
import matplotlib.pyplot as plt
```

```python
vt = 0.026
il = 10e-3
io = 100e-12
voc = 0.478
```

```python
tmp = []
#Method 1
v = voc
for i in range(0,120):
    v = vt*math.log((1+il/io)/(1+v/vt))
    tmp.append(v)
plt.plot(tmp)
plt.show()


```

```python
#Method 2
import numpy as np
tmp = []
v = voc
for i in range(0,20):
    v = vt*(il-io*(np.exp(v/vt)-1))/(io*(np.exp(v/vt)))
    tmp.append(v)
plt.plot(tmp)
```
